## 0.4 Funções Auxiliares

```python
# Função que cria etiquetas para gráficos
def create_labels(dataframe, column_name_str, column_name_number=None):
    if column_name_number != None:
        labels = [f"{dataframe[column_name_str][i]} - {dataframe[column_name_number][i]}" for i in dataframe.index]
    else:
        labels = [dataframe[column_name_str][i] for i in dataframe.index]
    return labels

# Função que cria gráficos de rosca
def generate_pie_plt(dataframe, column_name_str, column_name_number):
    plt.pie(dataframe[column_name_number], labeldistance = 1, radius = 3, wedgeprops = dict(width = 0.9))
    plt.legend(labels = create_labels(dataframe, column_name_str, column_name_number), loc = 'center')
    plt.title("Distribuição de Países", fontdict = {'fontsize':20,'fontweight':20})
    plt.show()
    
# Função que cria gráficos de coluna
def generate_bar_plot(dataframe, x_axis, y_axis, x_label, y_label, chart_title):
    plt.figure(figsize = (16,8))
    sns.barplot(data=dataframe, x = dataframe[x_axis], y = dataframe[y_axis])
    plt.ylabel(y_label)             
    plt.xlabel(x_label)
    plt.title(chart_title)
    plt.show()
    
# Função que cria dataframe com frequência de ocorrências de valores
def create_df_frequency(len_series_types, len_tv_show, frequency):
    data_tv_show = {"total_types": len_series_types, "total_tv_show": len_tv_show, "tv_show_frequency (%)": frequency}
    df = pd.DataFrame(data=data_tv_show, index=[0])
    return df

# Função que calcula frequência da ocorrência de valores
def calculate_frequency(series, type_seires):
    len_series = len(series)
    len_tv_show = len(series[series == type_seires])
    frequency = round((len_tv_show/len_series*100),2)
    return create_df_frequency(len_series, len_tv_show, frequency)
```

## 1.0 Analisando Quantidade de atores e atrizes

### 1.1 Os top 10 atores e atrizes considerando a base de dados unificada (amazon + netflix)

```python
# Selecionando os top 10 atores e atrizes da base unificada 
actors = merged_titles_df["cast"].dropna()
list_names = []

for actor in actors:
    list_actors = actor.split(",")
    for name in list_actors:
        if not name.isnumeric():
            list_names.append(name)
names_counter = Counter(list_names)
top_ten_actors = pd.DataFrame(names_counter.most_common(10), columns=["name", "count"])
```

```python
# Criando gráfico que exibe a quantidade de filmes com participação dos top 10 atores e atrizes
generate_pie_plt(top_ten_actors, "name", "count")
```

## 2.0 Analisando os top 5 países produtores de conteúdo

### 2.1 Os top 5 países produtores de conteúdo considerando a base unificada.

```python
# Selecionando os top 5 produtores de conteúdo da base unificada
countries = merged_titles_df["country"].dropna()
five_countries_general = pd.DataFrame(countries.value_counts().head()).reset_index()

# Renomeando colunas
five_countries_general = five_countries_general.rename(columns={"index": "country", "country": "count"})
```

```python
# Criando gráfico que exibe a distribuição de filmes por países da base unificada
generate_pie_plt(five_countries_general, "country", "count")
```

### 2.2.1 Os top 5 países produtores de conteúdo da Netflix

```python
# Selecionando os top 5 produtores da netflix
netflix_countries = netflix_df["country"].dropna()
five_countries_netflix = pd.DataFrame(netflix_countries.value_counts().head()).reset_index()
five_countries_netflix = five_countries_netflix.rename(columns={"country":"count", "index":"country"})
```

```python
# Criando gráfico de barras que exibe a quantidade de filmes por países da Netflix
generate_bar_plot(five_countries_netflix,"country", "count", "Países", "Total de Conteúdo Produzido", 
                  "Top 5 Países Produtores de Conteúdo na Netflix")
```

## 3.0 Analisando mês com mais produções da Netflix

```python
# Selecionando todos os meses de produção da netflix
dates_added = netflix_df["date_added"].dropna()
dates_str = dates_added.to_string().replace(",", "")
dates_list = dates_str.split()
dates_list = [
    date for date in dates_list 
    if not date.isnumeric()   
]

# Contando as produções por mês
dates_counter = Counter(dates_list)

# Exibindo mês de maior produção
most_common_month = pd.DataFrame(dates_counter.most_common(1), columns=["name", "count"])
most_common_month
```

### 4.2.1 Quantidade de filmes puramente de comédia e de comédia com outros gêneros na base unificada

```python
# Selecionando filmes de comédia e de comédia combinada com outros gêneros
movie_types = merged_titles_df["listed_in"]
list_comedy_total = []
for movie_type in movie_types:
    if "," in movie_type:
        for gen in movie_type.split(","):
            if gen == "Comedy":
                list_comedy_total.append(gen)
    elif movie_type == "Comedy":
        list_comedy_total.append(movie_type)
        
# Exibindo quantidade de filmes de comédia
quantity_comedy_total = len(list_comedy_total)
comedy_total_df = {"listed_in": ["Comedy"], "total": [quantity_comedy_total]}
pd.DataFrame(data=comedy_total_df, index=[0])
```

### 4.2.2 Quantidade de filmes puramente de comédia na base unificada

```python
# Selecionando filmes apenas de comédia
list_comedy = []
for movie_type in movie_types:
    if movie_type == "Comedy":
        list_comedy.append(movie_type)
quantity_comedy = len(list_comedy)
only_comedy_df = {"listed_in": ["Comedy"], "total": [quantity_comedy]}
pd.DataFrame(data=only_comedy_df, index=[0])
```
